This notebook is used to retrieve data from wikipedia in order to build our corpus of documents that will be used in our evauation.

# Import libraries

```python
import wikipedia
```

We make a list of all the wikipedia pages we want to retrieve

```python
retrieve_list = ["Alan Turing (person)", "Charles Babbage (person)", "Ada Lovelace (person)", "Machine learning (science)", "artificial intelligence (science)", "Cryptography", "theoretical computer science (science)", "computer (machine)", "programmer (computer)", "agriculture"]
```

```python
list_wikipedia_page = []
for retrieve_string in retrieve_list:
    wikipedia_page = wikipedia.page(retrieve_string)
    list_wikipedia_page.append(wikipedia_page.content)
    with open(f"sample_data/{retrieve_string}.txt", 'w', encoding='utf-8') as f:
        f.write(wikipedia_page.content)
```
